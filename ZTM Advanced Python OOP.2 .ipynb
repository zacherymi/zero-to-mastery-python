```python
# Object Oriented Programming

#Class definition
class PlayerCharacter:
  #Class Object Attribute | Static
  membership = True
  
  def __init__(self, name,age):
    self.name = name 
    self.age = age

  def shout(self):
     print(f'My name is {self.name}')
    

player1 = PlayerCharacter('Gavin Sperling',29)
player2 = PlayerCharacter('Lord Zachery Mitchell of the Iron Throne',28)

print(player1.name)
print(player1.age)
print(player1.shout())
```

```python
class Cat:
  species  = 'mammal'
  def __init__(self, name, age):
    self.name = name
    self.age = age

cudi = Cat('Cudi', 7)
gigi = Cat('Gigi', 5)
ramone = Cat('Ramone', 3)

def get_oldest_cat(*args):
  return max(args)

print(f'The oldest cat is {get_oldest_cat(cudi.age, gigi.age, ramone.age)} years old.')
```
