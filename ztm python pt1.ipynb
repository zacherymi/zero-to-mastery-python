List Methods

```
basket = [1,2,3,4,5]

#Adds an item to the end of the list
new_list = basket.append(6)
#Appending doesn't produce a new list example:
print(new_list)
#Fix 
new_list = basket
print(new_list)

#Inserting 7 at the 6th index
basket.insert(6,7)
print(basket)

#Extend to the end of a list
basket.extend([8,9])
print(basket)

#Remove an item at a given index... Default the end
basket.pop()
print(basket)

#Remove a certain value NOT index
basket.remove(6)
print(basket)
```

```
basket = ['a','b','c','d']

#Index: searches for a given value and returns it's index
print(basket.index('d'))

#Index: Can also search a list given limits
print(basket.index('b', 0, 3))

#Another way to find a value in a list... Returns True or False
print('b' in basket)

#Count the occurence of an item in a list
print(basket.count('b'))
```

```
basket = ['b', 'a', 'c', 'e', 'd' ]

#List method used to sort a list
basket.sort()
print(basket)

#Does not modify the list but returns a sorted list
print(sorted(basket))

#Reverse modifying sort
basket.sort()
basket.reverse()
print(basket)

#Unmodifying reverse sort
print(basket[::-1])

#Still prints the last from the reverse method
print(basket)
```

```
#Dictionary: Unordered key value pair
dictionary = {
    123 : [1, 50, 40],
     True : 'titties',
    'c' : True
}

#Using key in a dictionary to print a value
print(dictionary[123])
print(dictionary[True])
```

```
#Dictionary declaration
user = {
    'basket' : [1,2,3],
    'greet' : 'hello',
    (1,2) : True
}
 #.get method rerutns the value of a key that exists and None if it doesn't
print(user.get('sword'))
print(user.get('greet'))

#in method to check if a value is in a dictionary... Returns a bool
print('size' in user)

#items method// returns all key value pairs as tuples
print(user.items())

#user.clear() clears dictionary and creates empty dictionary

#Copy method
userTwo = user.copy()
print(userTwo)

#Pop remove method | removes key value pair and returns the value
print(user.pop('greet'))
print(user)

#  .popItem removes a key value pair from the dictionary

#Update method to add key value pair | returns none
print(user.update({'juno':'sword'}))
print(user)

```

```
#Tuples | immutable data structure | faster than a lists
my_tuple = (1,2,3,4,5)

#Accessing a tuple
print(my_tuple[2])

#in
print(5 in my_tuple)

#Tuple slicing
new_tuple = my_tuple[1:2]
#Returns item before 2nd value... Therefore only item 1
print(new_tuple)

#shorthand tuple declaration
a,b,c, *other  = (1,2,3,4,5,6,7,8,9)

print(a)
print(b)
print(c)
#Returns the rest of the tuple as a list 
print(other)

#Count Method
print(my_tuple.count(5))

#Index Method
print(my_tuple.index(3))

#Length Method
print(len(my_tuple))

```

```
#Sets | unordered collections of unique objects

my_set = {1,2,3,4,5,5}
print(my_set)

#adding to a set
my_set.add(6)
print(my_set)

#Cannot index a set must use in
print((1 in my_set))

#Unique digit length
print(len(my_set))

#Set to list
myListTuple = my_set.copy()
print(list(myListTuple))
```

```
#Set function that converts a list filled with duplicate values to a tuple of unique values
my_list = [1,1,2,2,3,3,4,4]
my_set = set(my_list)
print(my_set)
```
